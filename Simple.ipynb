# Simple Dividend Reinvestment Calculator

Projects the value of an investment over time, assuming dividends are automatically reinvested at the prevailing share price each year.

**Assumptions**
- Dividends are paid once per year, at year end, then immediately reinvested.
- The dividend yield and price growth rate are constant each year.
- No taxes or fees.

```python
# --- Inputs (edit these) ---
initial_investment = 10_000.0   # dollars invested up front
initial_share_price = 100.0     # price per share at year 0
dividend_yield = 0.04           # 4% annual dividend yield
annual_price_growth = 0.06      # 6% annual share price growth
years = 20                      # holding period
```

```python
def simulate_drip(initial_investment, initial_share_price,
                  dividend_yield, annual_price_growth, years):
    """Year-by-year DRIP simulation. Returns a list of dicts."""
    shares = initial_investment / initial_share_price
    price = initial_share_price
    rows = [{
        'year': 0,
        'price': price,
        'shares': shares,
        'dividend': 0.0,
        'shares_bought': 0.0,
        'value': shares * price,
    }]
    for year in range(1, years + 1):
        price *= (1 + annual_price_growth)
        dividend_per_share = price * dividend_yield
        dividend_cash = shares * dividend_per_share
        shares_bought = dividend_cash / price
        shares += shares_bought
        rows.append({
            'year': year,
            'price': price,
            'shares': shares,
            'dividend': dividend_cash,
            'shares_bought': shares_bought,
            'value': shares * price,
        })
    return rows

results = simulate_drip(initial_investment, initial_share_price,
                        dividend_yield, annual_price_growth, years)
```

```python
# --- Year-by-year table ---
print(f"{'Year':>4} {'Price':>10} {'Shares':>12} {'Dividend':>12} {'Value':>14}")
print('-' * 56)
for r in results:
    print(f"{r['year']:>4} {r['price']:>10.2f} {r['shares']:>12.4f} "
          f"{r['dividend']:>12.2f} {r['value']:>14.2f}")
```

```python
# --- Summary ---
final = results[-1]
total_dividends = sum(r['dividend'] for r in results)
total_return = (final['value'] / initial_investment) - 1
cagr = (final['value'] / initial_investment) ** (1 / years) - 1

print(f"Initial investment:   ${initial_investment:,.2f}")
print(f"Final value:          ${final['value']:,.2f}")
print(f"Total dividends paid: ${total_dividends:,.2f}")
print(f"Final share count:    {final['shares']:,.4f}")
print(f"Total return:         {total_return:.2%}")
print(f"CAGR:                 {cagr:.2%}")
```
